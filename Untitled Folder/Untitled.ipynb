```python
from tqdm import tqdm
```

```python
pregraph = dd(dict)

for i in tqdm(range(len(fifteen))):
    rent_from = fifteen.loc[i]['대여소번호']
    rent_to =   fifteen.loc[i]['반납대여소번호']
    if rent_to not in pregraph[rent_from]:
        pregraph[rent_from][rent_to] = 1
    else:
        pregraph[rent_from][rent_to] +=1
```

```python
G = nx.DiGraph()

for from_key in pregraph.keys():
    current = pregraph[from_key]
    for to_key in current:
        weight = current[to_key]
        G.add_edge(from_key, to_key, weight = weight)
```

```python
nx.draw_spring(G, with_labels = True)
```

```python
pageranked = nx.pagerank(G, alpha = 0.8)
```

```python
res = sorted(pageranked.items(), key=(lambda x: x[1]), reverse=True)
```
