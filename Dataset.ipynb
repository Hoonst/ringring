# 자전거 대여 정보

```python
five  = pd.read_excel('dataset/rent/대여소별 대여내역(2015년).xlsx')
six   = pd.read_excel('dataset/rent/대여소별 대여내역(2016년).xlsx')
seven = pd.read_excel('dataset/rent/대여소별 대여내역(2017년).xlsx')
```

```python
eight_nine3 = pd.read_csv('dataset/rent/서울특별시_공공자전거 일별 대여건수_(2018_2019.03).csv')
nine_45 = pd.read_excel("dataset/rent/서울특별시 공공자전거 일별 대여건수(2019.4_5).xlsx")
nine_611 = pd.read_excel("dataset/rent/서울특별시 공공자전거 일별 대여건수_20190601_20191130.xlsx")
nine12_zero6 = pd.read_csv("dataset/rent/서울특별시 공공자전거 일별 대여건수_2010.12.01~2020.06.30.csv")
```

```python
five_sum = five.groupby([five['대여 일자 / 월']]).sum().reset_index()
six_sum = six.groupby([six['대여 일자 / 월']]).sum().reset_index()
seven_sum = seven.groupby([seven['대여 일자 / 월']]).sum().reset_index()
```

```python
five_sum.columns = ['대여일시', '대여건수']
six_sum.columns = ['대여일시', '대여건수']
seven_sum.columns = ['대여일시', '대여건수']

eight_nine3.columns = ['대여일시', '대여건수']
nine_45.columns = ['대여일시', '대여건수']
nine_611.columns = ['대여일시', '대여건수']
nine12_zero6.columns = ['대여일시', '대여건수']
```

```python
in_total = five_sum.append([six_sum, seven_sum, eight_nine3, nine_45, nine_611, nine12_zero6]).reset_index(drop=True)

in_total.columns = ['date', 'count']
in_total['date'] = pd.to_datetime(in_total['date'], format='%Y-%m-%d')
```

```python
in_total['count']
```

```python
in_total['count'] = in_total['count'].astype(float)
```

```python
in_total
```

```python
in_total.to_csv('in_total.csv', index=False, encoding='utf-8')
```

```python
def plot_line(df, title, x_axis = 'date', y_axis = 'count'):
    fig = px.line(df, x =x_axis, y =y_axis, title = title)
    fig.show()
```

```python
plot_line(in_total, '따릉이 대여(2015.09.19 ~ 2020.06.30)')
```
